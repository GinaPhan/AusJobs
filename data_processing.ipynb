### HQ State consistency

```python
import pandas as pd

# Define a dictionary to map abbreviations and variations to full names
state_mapping = {
    # Australia
    'NSW': 'New South Wales',
    'VIC': 'Victoria',
    'QLD': 'Queensland',
    'WA': 'Western Australia',
    'SA': 'South Australia',
    'TAS': 'Tasmania',
    'NT': 'Northern Territory',
    'ACT': 'Australian Capital Territory',
    
    # United States
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas',
    'CA': 'California', 'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware',
    'FL': 'Florida', 'GA': 'Georgia', 'HI': 'Hawaii', 'ID': 'Idaho',
    'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa', 'KS': 'Kansas',
    'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi',
    'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada',
    'NH': 'New Hampshire', 'NJ': 'New Jersey', 'NM': 'New Mexico', 'NY': 'New York',
    'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma',
    'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah',
    'VT': 'Vermont', 'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia',
    'WI': 'Wisconsin', 'WY': 'Wyoming',
    
    # Canada
    'AB': 'Alberta', 'BC': 'British Columbia', 'MB': 'Manitoba', 'NB': 'New Brunswick',
    'NL': 'Newfoundland and Labrador', 'NS': 'Nova Scotia', 'ON': 'Ontario',
    'PE': 'Prince Edward Island', 'QC': 'Quebec', 'SK': 'Saskatchewan',
}

# Function to standardize state names
def standardize_state(state):
    if pd.isna(state):
        return state
    state = str(state).strip()
    # Check if the state is already in the correct format
    if state in state_mapping.values():
        return state
    # Check if it's an abbreviation or variation
    return state_mapping.get(state.upper(), state)

filename = "company_information_full.xlsx"
df = pd.read_excel(filename)

# Apply the standardization to the 'hq_state' column
df['hq_state'] = df['hq_state'].apply(standardize_state)

# Save the cleaned data
df.to_excel('cleaned_state_data.xlsx', index=False)

print("Data cleaning completed. Cleaned data saved to 'cleaned_state_data.xlsx'.")

print("\nUnique values in hq_state after cleaning:")
print(df['hq_state'].value_counts())

# Check for any remaining potential inconsistencies
print("\nPotential remaining inconsistencies:")
print(df['hq_state'][~df['hq_state'].isin(state_mapping.values())].unique())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify

def prepare_data(df_stated):
    # Count occurrences of each company type
    type_counts = df_stated['company_type'].value_counts()

    # Create a DataFrame for the analysis
    df_analysis = pd.DataFrame({'count': type_counts})
    df_analysis['percentage'] = df_analysis['count'] / len(df_stated) * 100

    # Group the data
    def group_company_types(company_type):
        if company_type in ['PRIVATELY_HELD', 'SELF_OWNED']:
            return 'Private'
        elif company_type == 'PUBLIC_COMPANY':
            return 'Public'
        elif company_type == 'PARTNERSHIP':
            return 'Partnership'
        else:
            return 'Other'

    df_analysis['group'] = df_analysis.index.map(group_company_types)

    # Calculate group totals
    group_totals = df_analysis.groupby('group')['count'].sum().sort_values(ascending=False)
    group_percentages = group_totals / len(df_stated) * 100

    return df_analysis, group_totals, group_percentages

def print_analysis(df_analysis, group_totals, group_percentages):
    print("Company Type Analysis:")
    print(df_analysis)
    print("\nTotal companies:", df_analysis['count'].sum())
    print("\nGrouped Company Types:")
    print(group_totals)
    print("\nGroup Percentages:")
    print(group_percentages)

def plot_vertical_bar(df_analysis):
    plt.figure(figsize=(12, 6))
    df_analysis['percentage'].plot(kind='bar')
    plt.title('Distribution of Company Types')
    plt.xlabel('Company Type')
    plt.ylabel('Percentage')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

def plot_horizontal_bar(df_analysis):
    plt.figure(figsize=(12, 6))
    df_analysis['percentage'].sort_values().plot(kind='barh')
    plt.title('Distribution of Company Types (Horizontal)')
    plt.xlabel('Percentage')
    plt.ylabel('Company Type')
    plt.tight_layout()
    plt.show()

def plot_pie_chart(group_percentages):
    plt.figure(figsize=(10, 10))
    group_percentages.plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title('Distribution of Grouped Company Types')
    plt.ylabel('')
    plt.show()

def plot_treemap(group_totals):
    plt.figure(figsize=(12, 8))
    squarify.plot(sizes=group_totals.values, label=group_totals.index, 
                  alpha=.8, text_kwargs={'fontsize':12})
    plt.title('Treemap of Grouped Company Types')
    plt.axis('off')
    plt.show()

def plot_donut_chart(group_percentages):
    plt.figure(figsize=(10, 10))
    plt.pie(group_percentages, labels=group_percentages.index, autopct='%1.1f%%', 
            startangle=90, pctdistance=0.85, wedgeprops=dict(width=0.5))
    plt.title('Donut Chart of Grouped Company Types')
    plt.show()

def plot_heatmap(df_analysis):
    plt.figure(figsize=(12, 4))
    sns.heatmap(df_analysis['percentage'].values.reshape(1, -1), 
                annot=True, fmt='.1f', cmap='YlOrRd', 
                xticklabels=df_analysis.index, yticklabels=['Percentage'])
    plt.title('Heatmap of Company Type Percentages')
    plt.xlabel('Company Type')
    plt.tight_layout()
    plt.show()

def plot_grouped_bar(df_analysis):
    plt.figure(figsize=(12, 6))
    df_analysis.groupby('group')['percentage'].sum().plot(kind='bar')
    plt.title('Percentage of Company Types by Group')
    plt.xlabel('Group')
    plt.ylabel('Percentage')
    plt.show()

# Main execution
if __name__ == "__main__":
    # Assuming df_stated is your DataFrame and it has a column named 'company_type'
    # If you're running this in a Jupyter notebook, you may need to define df_stated before this point

    df_analysis, group_totals, group_percentages = prepare_data(df_stated)
    
    print_analysis(df_analysis, group_totals, group_percentages)
    
    # Generate all charts
    plot_vertical_bar(df_analysis)
    plot_horizontal_bar(df_analysis)
    plot_pie_chart(group_percentages)
    plot_treemap(group_totals)
    plot_donut_chart(group_percentages)
    plot_heatmap(df_analysis)
    plot_grouped_bar(df_analysis)
```
